```python
model_name = "wentao-yuan/robopoint-v1-vicuna-v1.5-13b"
model, processor, tokenizer = load_model(model_name)
```

```python
from PIL import Image
import math

def plot_scatter(logits, distances, title="Scatterplot"):
    with Image.open(image_path) as image:
        
        # Create a scatter plot
        plt.scatter(logits,distances)
        
        # Adding labels
        plt.xlabel('X-axis Label')
        plt.ylabel('Y-axis Label')
        plt.title(title)
        
        # Show the plot
        plt.show()
        plt.close()

def calculate_euclidian_distance(coord1, coord2):
    x1, y1 = coord1
    x2, y2 = coord2
    distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return distance
```

```python
tasks, images, ground_truths, object_list = extract_all("../data/first_experiments")

NUMBER_OF_RUNS = 5

for i in range(len(tasks)):   
    print(f"Task: {i + 1}")
    print(tasks[i])

    image = images[i]
    with Image.open(image) as img:
        width, height = img.size
    task = tasks[i]
    objects = object_list[i]
    groundtruths = ground_truths[i]


    for object, groundtruth in zip(objects, groundtruths):
        distances = []
        probs = []
        prompt = generate_prompt(object, "Locate several points within the vacant space of the ")
        for j in range(NUMBER_OF_RUNS):
            generated_text, outputs = do_inference_with_logits(image, prompt, model, processor, tokenizer, 0.2)
            first_coordinate = get_coordinates(generated_text, width, height)[0]
            first_prob = calculate_probs_per_coordinate(outputs)[0]
            distance = calculate_euclidian_distance(first_coordinate, groundtruth)

            distances.append(distance)
            probs.append(first_prob)
        
            print(f"Run: {j + 1} - Output: {generated_text}")
            print(f"Probs: {probs}")
            
        plot_scatter(probs, distances, object)
```

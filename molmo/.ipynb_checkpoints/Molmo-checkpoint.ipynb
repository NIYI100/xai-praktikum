```python
from transformers import AutoModelForCausalLM, AutoProcessor, GenerationConfig
from PIL import Image
import requests
import torch
import os
import matplotlib.pyplot as plt
import re
```

```python
model_name = 'allenai/Molmo-7B-D-0924'
```

```python
# load the processor
processor = AutoProcessor.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype='auto',
    device_map='auto'
)
```

```python
# load the model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    #torch_dtype='auto',
    torch_dtype=torch.float16,
    device_map='auto'
)
```

```python
def iterate_directory():
    tasks = []
    images = []
    for example in os.listdir("./data"):
        if example.startswith("."):
            continue
        subdir_path = os.path.join("./data", example)
        lang_file_path = os.path.join(subdir_path, "lang.txt")
        
        with open(lang_file_path, "r") as lang_file:
            task = lang_file.read()
            tasks.append(task)

        for file in os.listdir(subdir_path):
                    if file.startswith("im_") and file.endswith((".png", ".jpg", ".jpeg")):
                        image_path = os.path.join(subdir_path, file)
                        image = Image.open(image_path)
                        images.append(image)

    return tasks, images        
```

```python
def parse_coordinates(string):
    # Regular expression to match object name and coordinates
    pattern = r'\(([^,]+),\s*\((\d+),\s*(\d+)\)\)'
    matches = re.findall(pattern, string)
    
    coordinates = []
    for obj, x, y in matches:
        coordinates.append((obj.strip(), int(x), int(y)))
    return coordinates


def scale_coordinates(coordinates, image_width, image_height):
    scaled_coords = []
    for obj, x, y in coordinates:
        # Scale coordinates to the image size
        scaled_x = x / 100 * image_width
        scaled_y = y / 100 * image_height
        scaled_coords.append((obj, int(scaled_x), int(scaled_y)))
    return scaled_coords
```

```python
def visualize_points_on_image(image, coordinates):
    # Load the image
    plt.imshow(image)
    image_width, image_height = image.size

    # Scale coordinates to image dimensions
    scaled_coordinates = scale_coordinates(coordinates, image_width, image_height)
    
    # Extract and plot the points
    x_coords = [coord[1] for coord in scaled_coordinates]
    y_coords = [coord[2] for coord in scaled_coordinates]
    plt.scatter(x_coords, y_coords, c='red', marker='o')
    
    # Annotate each point with its object name
    for obj, x, y in scaled_coordinates:
        plt.text(x, y, obj, color='blue', fontsize=9, ha='right', va='bottom')
    
    # Add labels and show the plot
    plt.title("Coordinates on Image")
    plt.axis("on")  # Show axes
    plt.show()
```

```python
prompt_start = "You are the robot in the picture. Your task is to: "
prompt_1 = "To complete this task, first find the relevant objects in the picture. Afterwards create a plan to execute"
prompt_end = " Your output format should be like this: (name of object 1, (x1, y1)), (name of object 2, (x2, y2)), ..., with x and y beeing integer values"

#prompt_1 = "You are the robot in the picture. Your task is to: put cup from counter or drying rack into sink. Find the relevant objects for that in the picture and output the coordinates of these objects."
```

```python
tasks, images = iterate_directory()

for i in range(len(tasks)):
    print(f"Task: {i + 1}")

    # Process inputs
    inputs = processor.process(
        images=images[i],
        text=prompt_start + tasks[i] + prompt_1 + prompt_end
    )
    inputs = {k: v.to(model.device).unsqueeze(0) for k, v in inputs.items()}
    
    # generate output
    with torch.autocast(device_type="cuda", enabled=True, dtype=torch.bfloat16):
      output = model.generate_from_batch(
          inputs,
          GenerationConfig(max_new_tokens=200, stop_strings="<|endoftext|>"),
          tokenizer=processor.tokenizer
      )
    generated_tokens = output[0,inputs['input_ids'].size(1):]
    generated_text = processor.tokenizer.decode(generated_tokens, skip_special_tokens=True)

    coordinates = parse_coordinates(generated_text)
    visualize_points_on_image(images[i], coordinates)
    
    print(generated_text)
    print("")
    print("")
```
